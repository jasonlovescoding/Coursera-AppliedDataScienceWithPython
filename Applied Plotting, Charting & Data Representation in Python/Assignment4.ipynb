### Records of TOP 4 Clubs in 2016 J1 League from 1996 to 2016

```python
import pandas as pd
import numpy as np
```

```python
# 4th: Gamba Osaka
url = 'https://en.wikipedia.org/wiki/Gamba_Osaka'
Gamba_Osaka = pd.read_html(url)[3]
Gamba_Osaka = Gamba_Osaka.rename(columns = Gamba_Osaka.iloc[0]).iloc[1:]
Gamba_Osaka.set_index('Season', inplace = True)
Gamba_Osaka.head()
```

```python
Gamba_Osaka_pos = []
for idx, row in Gamba_Osaka.iterrows():
    if idx >= '1996' and idx <= '2016':
        if row['Div.'] == 'J1':
            Gamba_Osaka_pos.append(int(row['Pos.']))
        else:
            Gamba_Osaka_pos.append(np.nan)
print(Gamba_Osaka_pos)
```

```python
# 3th: Kashima Antlers
url = 'https://en.wikipedia.org/wiki/Kashima_Antlers'
Kashima_Antlers = pd.read_html(url)[9]
Kashima_Antlers = Kashima_Antlers.rename(columns = Kashima_Antlers.iloc[0]).iloc[1:]
Kashima_Antlers.set_index('Season', inplace = True)
Kashima_Antlers.head()
```

```python
Kashima_Antlers_pos = []
for idx, row in Kashima_Antlers.iterrows():
    if idx >= '1996' and idx <= '2016':
        if row['Division'] == 'J1':
            Kashima_Antlers_pos.append(int(row['Position']))
        else:
            Kashima_Antlers_pos.append(np.nan)
print(Kashima_Antlers_pos)
```

```python
# 2nd: Kawasaki Frontale
url = 'https://en.wikipedia.org/wiki/List_of_Kawasaki_Frontale_records_and_statistics'
Kawasaki_Frontale = pd.read_html(url)[1]
Kawasaki_Frontale = pd.concat([Kawasaki_Frontale.ix[1:2, 0:8], Kawasaki_Frontale.ix[5:, 0:8]])
for i in range(8, 1, -1):
    Kawasaki_Frontale.loc[5][i] = Kawasaki_Frontale.loc[5][i-1] 
Kawasaki_Frontale.loc[5][0] = '2000'
Kawasaki_Frontale.loc[5][1] = 'J1'
Kawasaki_Frontale = Kawasaki_Frontale.rename(columns = Kawasaki_Frontale.iloc[0]).iloc[1:]
Kawasaki_Frontale.set_index('Season', inplace = True)
Kawasaki_Frontale = Kawasaki_Frontale[:-2]
Kawasaki_Frontale.head()
```

```python
import re
Kawasaki_Frontale_pos = [np.nan, np.nan, np.nan]
for idx, row in Kawasaki_Frontale.iterrows():
        if row['League'] == 'J1':
            matcher = re.compile('(\d*) / (\d*)')
            matchObj = matcher.match(row['Place'])
            Kawasaki_Frontale_pos.append(int(matchObj.group(1)))
        else:
            Kawasaki_Frontale_pos.append(np.nan)
Kawasaki_Frontale_pos.append(3)
print(Kawasaki_Frontale_pos)
```

```python
# 1st: Urawa Red Diamonds
url = 'https://en.wikipedia.org/wiki/Urawa_Red_Diamonds'
Urawa_Red_Diamonds = pd.read_html(url)[3]
Urawa_Red_Diamonds = Urawa_Red_Diamonds.rename(columns = Urawa_Red_Diamonds.iloc[0]).iloc[1:]
Urawa_Red_Diamonds.set_index('Season', inplace = True)
Urawa_Red_Diamonds.head()
```

```python
Urawa_Red_Diamonds_pos = []
for idx, row in Urawa_Red_Diamonds.iterrows():
    if idx >= '1996' and idx <= '2016':
        if row['Div.'] == 'J1':
            Urawa_Red_Diamonds_pos.append(int(row['Pos.']))
        else:
            Urawa_Red_Diamonds_pos.append(np.nan)
print(Urawa_Red_Diamonds_pos)
```

```python
pos = pd.DataFrame({'Gamba Osaka': Gamba_Osaka_pos,
                    'Kashima Antlers': Kashima_Antlers_pos,
                    'Kawasaki Frontale': Kawasaki_Frontale_pos,
                    'Urawa Red Diamonds': Urawa_Red_Diamonds_pos},
                    index = np.arange(1996, 2017))
# 19 means relegated
pos.fillna(19, inplace = True)
pos.head()
```
