```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

%matplotlib notebook

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
samp_size = 1000
x1 = np.random.normal(0, 1, samp_size)
x2 = np.random.gamma(1, 1, samp_size)
x3 = np.random.exponential(1, samp_size)
x4 = np.random.uniform(0, 1, samp_size)
xs = [x1, x2, x3, x4]

# plot the histograms
gspec = gridspec.GridSpec(2, 2)
gspec.update(hspace=0.4)
axs = [plt.subplot(gspec[0, 0]), plt.subplot(gspec[0, 1]), plt.subplot(gspec[1, 0]), plt.subplot(gspec[1, 1])]
color = ['blue', 'orange', 'green', 'red']
texts = ['x1 Normal', 'x2 Gamma', 'x3 Exponential', 'x4 Uniform']
for n in range(len(axs)):
    axs[n].hist(xs[n], color = color[n], alpha = 0.5, normed = True, bins = 25, edgecolor = 'none')
    axs[n].tick_params(bottom='off', left='off', top = 'off', right = 'off', axis = 'both', labelsize = 9)
    for spine in axs[n].spines.values():
        spine.set_visible(False)
    axs[n].set_title(texts[n], fontsize=11)
```
